**Question 1:**

**What is hypothesis testing in statistics?**

-- **Answer**
- Hypothesis testing is a method to check if the results of a sample are true for the whole population. It helps decide whether to accept or reject a claim (hypothesis) based on data.
---
**Question 2:**

**What is the null hypothesis, and how does it differ from the alternative hypothesis?**

-- **Answer**
- The null hypothesis says there is no effect or no difference.
The alternative hypothesis says there is an effect or difference.
We test data to see if we should reject the null in favor of the alternative.

---
**Question 3:**

**Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.**

-- **Answer**
- The significance level is the chance of rejecting the null hypothesis when it is actually true. Commonly, if it's equal to 0.05. If the p-value < significance level, we reject the null hypothesis.

---
**Question 4:**

**What are Type I and Type II errors? Give examples of each.**

-- **Answer**
- Type I error: Rejecting a true null hypothesis. Example: Saying a medicine works when it doesn’t.

- Type II error: Not rejecting a false null hypothesis. Example: Saying a medicine doesn’t work when it actually does.

---
**Question 5:**

**What is the difference between a Z-test and a T-test? Explain when to use
each.**

-- **Answer**
- Z-test: Used when the population variance is known and the sample size is large (n > 30).

- T-test: Used when the population variance is unknown and the sample size is small (n < 30).

**Question 7:**

**Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

**Answer:**

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]
mu_0 = 50.0
alpha = 0.05

z_stat, p_value = ztest(sample_data, value=mu_0, alternative='two-sided')

print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Result: Reject H0")
else:
    print("Result: Fail to reject H0")
```

**Question 8:**

**Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(loc=50, scale=10, size=100)

ci = stats.t.interval(0.95, len(data)-1, loc=np.mean(data), scale=stats.sem(data))
print("95% CI:", ci)

plt.hist(data, bins=15, alpha=0.7)
plt.axvline(np.mean(data), color='red', linestyle='dashed', label='Mean')
plt.legend()
plt.show()
```

**Question 9:**

**Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.**

**Answer:**
- A Z-score represents how many standard deviations a specific data point is away from the mean of the dataset.

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

sample_data = np.random.normal(loc=50, scale=5, size=100)

z_scores = st.zscore(sample_data)

print(f"Mean of Z-scores: {np.mean(z_scores)}")
print(f"Std Dev of Z-scores: {np.std(z_scores)}")

plt.figure(figsize=(8, 5))
plt.hist(z_scores, bins=10, alpha=0.7)
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')


plt.axvline(z_scores.mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
