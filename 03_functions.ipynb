10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - map() - it executes specified function for each element or item of the provided iterable.
  reduce() - this function helps in reduction, it reduces the items to single value using specified function on iterable.
  filter() - it basically filters the items based on the condition of the function.

```python
from functools import reduce    # for reduce function

nums = [1, 2, 3, 4]

print(list(map(lambda x: x+1, nums))) # map function
print(reduce(lambda x, y: x + y, nums)) # reduce function
print(list(filter(lambda x: x%2==0, nums))) # filter function
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.






```python
def even_sum(l1):
  sum = 0
  for i in l1:
    if(i%2 == 0):
      sum += i
  return sum

l1 = [2,3,4,1,6,8,7,9]
print("sum : ",even_sum(l1))
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.



```python
def squared(l1):
  l2 = []
  for i in l1:
    sq = i * i
    l2.append(sq)
  return l2

l1 = [1,2,3,4,5]
print(squared(l1))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.




```python
def is_prime(n):
  if n < 2:
    return "Not Prime"
  else:
    for i in range(2,n):
      if(n%i == 0):
        return "Not Prime"
  return "Prime"

n = 113

print(is_prime(n))
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.



```python
def fibo_gen(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

n = int(input("Enter number of terms: "))
for i in fibo_gen(n):
    print(i)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.



```python
def gen(n):
  for i in range(n):
    yield 2**i

n = 5
g = gen(5)
for i in g:
  print(i)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.



```python
def read_file(f_name):
    with open(f_name, 'r') as f:
        for i in f:
            yield i

for i in read_file('/content/sample.txt'):
  print(i)
```
